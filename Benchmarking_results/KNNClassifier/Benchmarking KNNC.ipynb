# Becnhmarking

```python
def benchmark_time(n_neighbours_i, metric_i, x_i, y_i, x_val):
    time_taken = 0

    for _ in range(5):
        # training the model
        start_time = time()

        if metric_i != 'minkowski':
            model = KNeighborsClassifier(n_neighbors=n_neighbours_i,        # k
                                         metric=metric_i,
                                         )

        else:
            model = KNeighborsClassifier(n_neighbors=n_neighbours_i,        # k
                                         metric=metric_i,
                                         p=3
                                         )

        # fit the dataset into the model
        model.fit(x_i, y_i)

        # use the model to predict the data
        _ = model.predict(x_val)

        end_time = time()
        time_taken += (end_time - start_time)

    return time_taken / 5
```

```python
from sklearn.neighbors import KNeighborsClassifier
from time import time
import pandas as pd

metric_choices = ['manhattan', 'euclidean', 'minkowski']
n_neighbours_choices = [23, 32, 71, 100, 224, 316, 374]
datasets = ['cs-500.csv', 'cs-1000.csv', 'cs-5000.csv', 'cs-10000.csv', 'cs-50000.csv', 'cs-100000.csv', 'cs-140000.csv']
val_dataset = pd.read_csv("cs_val_2.csv")
x_val = val_dataset.drop(columns=['target'])

for metric in metric_choices:
    time_taken = []

    for i in range(len(datasets)):
        df = pd.read_csv(datasets[i])
        x = df.drop(columns=['target'])
        y = df['target']

        time_here = benchmark_time(n_neighbours_choices[i], metric, x, y, x_val)
        time_taken.append(time_here)

    print(f"\nTime taken for {metric} distance: ")
    print(time_taken)
```
